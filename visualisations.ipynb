```python
fig=plt.figure(figsize=(6,4),facecolor='green',edgecolor='red',linewidth=20)
x=np.linspace(0,10,100)
y=np.sin(x)
plt.plot(x,y,marker='o',markersize=15,markerfacecolor='blue',markeredgecolor='yellow',markeredgewidth=2)
```

```python
fig=plt.figure(figsize=(6,4),facecolor='green',edgecolor='red',linewidth=20)
x=np.linspace(0,10,100)
y=np.sin(x)
plt.plot(x,y,linestyle=':',color='red',linewidth=1)
```

```python
fig=plt.figure(figsize=(6,4),facecolor='green',edgecolor='red',linewidth=20)
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,linestyle=':',color='red',linewidth=1,label='sin(x)')
plt.plot(x,z,linestyle=':',color='blue',linewidth=1,label='cos(x)')
plt.title('time series')
plt.xlabel('time')
plt.ylabel('value')
plt.legend()
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,linestyle=':',color='red',linewidth=1,label='sin(x)')
plt.plot(x,z,linestyle=':',color='blue',linewidth=1,label='cos(x)')
plt.title('time series',loc='left',pad=10,fontdict={
                                                    'fontsize'==16,
                                                    'color'=='blue',
                                                    'weight'=='bold'
                                                    })
plt.xlabel('time')
plt.ylabel('value')
plt.legend()
```

```python
fig=plt.figure(figsize=(6,4),facecolor='green',edgecolor='red',linewidth=20)
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,linestyle=':',color='red',linewidth=1,label='sin(x)')
plt.plot(x,z,linestyle=':',color='blue',linewidth=1,label='cos(x)')
plt.title('time series')
plt.xlabel('time')
plt.ylabel('value')
plt.legend()
plt.xlim(2,6)
plt.ylim(-1,1)
```

```python
rainfall=[0.8,0.6,0,0,0.1,0.7,0.8,0.6,1.2,2.3,1.4,1.5]
months=['Jan','Feb','Mar','Apr',
        'May','Jun','July','Aug',
        'Sep','Oct','Nov','Dec']
fig=plt.figure(figsize=(6,4))
plt.bar(months,rainfall,color='blue',hatch='//')
plt.title('Rainfall in 2019')
plt.xlabel('Month')
plt.ylabel('Rainfall(mm)')
```
