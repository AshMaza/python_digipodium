```python
#removing unwanted columns
cols_to_delete=['_id','pid','url','images']
df.drop(cols_to_delete,axis=1,inplace=True)
#axis=1 vertically
#inplace=True , means update in the memory
```

```python
def clean_price(price):
    if price is None:
        return np.nan
    elif isinstance(price,str):
        price=price.replace(',','')
        if price:
            price=float(price)
        else:
            price=np.nan
        return price
    else:
        return price
    
df['actual_price']=df['actual_price'].apply(clean_price)
```

```python
def clean_discount(discount):
    if discount is None:
        return np.nan
    elif isinstance(discount,str):
        discount=discount.replace('% off','')
        if discount:
            discount=float(discount)
        else:
            discount=np.nan
        return discount
    else:
        return discount
    
df['discount']=df['discount'].apply(clean_discount)
```
