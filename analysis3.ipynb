```python
df.groupby('category')['selling_price'].mean()
```

```python
def clean_sp(sp):
    if sp is None:
        return np.nan
    elif isinstance(sp,str):
        sp=sp.replace(',','')
        if sp:
            sp=float(sp)
        else:
            sp=np.nan
        return sp
    else:
        return sp
    
df['selling_price']=df['selling_price'].apply(clean_sp)
df['actual_price']=df['actual_price'].apply(clean_sp)
```

```python
df.groupby('category')['selling_price'].mean()
```

```python
seller_data=df.groupby('seller')['selling_price'].mean().reset_index()
seller_data.sort_values('selling_price',ascending=False).head(25)
```
